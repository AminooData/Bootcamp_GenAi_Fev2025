```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Charger les donn√©es depuis le fichier CSV
df = pd.read_csv('2019.csv')
df.head()
```

```python
plt.figure(figsize=(10, 6))  # D√©finir la taille de la figure
plt.scatter(df['Social support'], df['Score'], color='blue', alpha=0.5)
plt.title('Relation entre le soutien social et le score de bonheur')
plt.xlabel('Soutien social')
plt.ylabel('Score de bonheur')
plt.grid(True)
plt.show()
```

```python

print(df.columns.tolist())
df.rename(columns={'Country or region': 'Region'}, inplace=True)
print(df['Region'])
```

```python
# D√©finir la taille de la figure
fig, ax1 = plt.subplots(figsize=(12, 8))

# Tracer le diagramme √† barres pour le PIB par habitant
bar_width = 0.4
bar_positions = range(len(df['Region']))
bars = ax1.bar(bar_positions, df['GDP per capita'], bar_width, align='center', color='skyblue', label='PIB par habitant')

# Configuration de l'axe des ordonn√©es pour le PIB
ax1.set_xlabel('R√©gion')
ax1.set_ylabel('PIB par habitant (‚Ç¨)', color='blue')
ax1.set_xticks(bar_positions)
ax1.set_xticklabels(df['Region'], rotation=45, ha='right')
ax1.tick_params(axis='y', labelcolor='blue')

# Cr√©ation d'un second axe des ordonn√©es pour l'esp√©rance de vie
ax2 = ax1.twinx()
ax2.plot(bar_positions, df['Healthy life expectancy'], color='green', marker='o', label='Esp√©rance de vie en bonne sant√©')
ax2.set_ylabel('Esp√©rance de vie en bonne sant√© (ann√©es)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Ajouter une l√©gende
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))

# Ajouter un titre
plt.title('Comparaison du PIB par habitant et de l\'esp√©rance de vie en bonne sant√© par r√©gion')

# Ajuster la mise en page pour √©viter le chevauchement
fig.tight_layout()

# Afficher le graphique
plt.show()
```
