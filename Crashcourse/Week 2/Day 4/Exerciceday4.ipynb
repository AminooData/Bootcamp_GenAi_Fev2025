```python
# Exercice 1: Comprendre la visualisation des données


# La visualisation des données est essentielle en analyse de données car elle permet de transformer des données brutes en représentations graphiques compréhensibles. Cela facilite l'identification de tendances, de modèles et d'anomalies qui pourraient ne pas être évidents dans des tableaux de chiffres. De plus, des visualisations claires et précises aident à communiquer efficacement les résultats et les insights aux parties prenantes, améliorant ainsi la prise de décision basée sur les données.

# Un graphique linéaire, ou graphique en ligne, est utilisé pour afficher des données continues sur une période de temps. Il permet de visualiser les tendances, les augmentations ou les diminutions, et les fluctuations au fil du temps. C'est particulièrement utile pour analyser des séries chronologiques, comme les variations de température, les ventes mensuelles ou les taux de croissance.
```

```python
# Exercice 2
import matplotlib.pyplot as plt

# Jours de la semaine
jours = ['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']
# Températures correspondantes
temperatures = [22, 23, 21, 20, 24, 25, 23]

# Création du graphique linéaire
plt.plot(jours, temperatures, marker='o', linestyle='-', color='b')

# Ajout des étiquettes et du titre
plt.xlabel('Jour')
plt.ylabel('Température (°C)')
plt.title('Variation de la température sur une semaine')

# Affichage du graphique
plt.show()
```

```python
# Exercice 3 
import matplotlib.pyplot as plt

# Mois de l'année
mois = ['Janvier', 'Février', 'Mars', 'Avril', 'Mai']
# Ventes correspondantes en euros
ventes = [5000, 5500, 6200, 7000, 7500]

# Création du graphique à barres
plt.bar(mois, ventes, color='green')

# Ajout des étiquettes et du titre
plt.xlabel('Mois')
plt.ylabel('Montant des ventes (€)')
plt.title('Ventes mensuelles du magasin de détail')

# Affichage du graphique
plt.show()
```

```python
# Exercice 4
import seaborn as sns
import matplotlib.pyplot as plt

# Supposons que 'df' est un DataFrame contenant une colonne 'ACPG_Catégorie'
# Exemple de création de données fictives
import pandas as pd
df = pd.DataFrame({
    'ACPG_Catégorie': ['A', 'B', 'A', 'C', 'B', 'A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C', 'A']
})

# Création de l'histogramme
sns.histplot(df['ACPG_Catégorie'], color='skyblue')

# Ajout du titre et des étiquettes
plt.title('Distribution des catégories d\'ACPG des étudiants')
plt.xlabel('Catégorie ACPG')
plt.ylabel('Fréquence')

# Affichage du graphique
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Supposons que 'df' est un DataFrame contenant les colonnes 'Anxiété' et 'Sexe'
# Exemple de création de données fictives
df = pd.DataFrame({
    'Anxiété': ['Oui', 'Non', 'Oui', 'Oui', 'Non', 'Non', 'Oui', 'Non', 'Oui', 'Non'],
    'Sexe': ['Féminin', 'Masculin', 'Féminin', 'Masculin', 'Féminin', 'Masculin', 'Féminin', 'Masculin', 'Féminin', 'Masculin']
})

# Création du graphique à barres
sns.countplot(x='Sexe', hue='Anxiété', data=df, palette='Set2')

# Ajout du titre et des étiquettes
plt.title('Comparaison des niveaux d\'anxiété entre les sexes')
plt.xlabel('Sexe')
plt.ylabel('Nombre d\'étudiants')

# Affichage du graphique
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Supposons que 'df' est un DataFrame contenant les colonnes 'Âge' et 'Crises_de_panique'
# Exemple de création de données fictives
df = pd.DataFrame({
    'Âge': [20, 22, 21, 23, 24, 22, 21, 23, 20, 24],
    'Crises_de_panique': ['Oui', 'Non', 'Oui', 'Non', 'Oui', 'Non', 'Oui', 'Non', 'Oui', 'Non']
})

# Conversion des réponses en valeurs numériques
df['Crises_de_panique_num'] = df['Crises_de_panique'].map({'Oui': 1, 'Non': 0})

# Création du nuage de points
sns.scatterplot(x='Âge', y='Crises_de_panique_num', data=df, s=100, color='red', marker='o')

# Personnalisation du graphique
plt.title('Relation entre l\'âge et les crises de panique')
plt.xlabel('Âge')
plt.ylabel('Crises de panique (Oui=1, Non=0)')
plt.yticks([0, 1], ['Non', 'Oui'])  # Remplace les valeurs 0 et 1 par 'Non' et 'Oui'

# Affichage du graphique
plt.show()
```
