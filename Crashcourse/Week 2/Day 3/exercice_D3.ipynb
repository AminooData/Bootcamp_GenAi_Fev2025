```python
# exercice 1 
import pandas as pd
df = pd.read_csv('train.csv')  # Remplace 'train.csv' par le chemin vers ton fichier
print("Nombre de lignes initial :", df.shape[0])

duplicates = df.duplicated()
print("Nombre de lignes dupliquées :", duplicates.sum())

df = df.drop_duplicates()
print("Nombre de lignes après suppression :", df.shape[0])

# EXERCICE 2 : Chargement du dataset Titanic et analyse initiale

import pandas as pd

# Charger le dataset Titanic 
df_titanic = pd.read_csv('train.csv')

# Afficher le nombre de lignes initial
print("Nombre de lignes initial :", df_titanic.shape[0])

# Afficher les 5 premières lignes pour avoir un aperçu du contenu
print("Premières lignes du dataset Titanic:")
print(df_titanic.head())

# Afficher les informations générales (nombres de valeurs non nulles et types de données)
print("Informations sur le dataset Titanic:")
print(df_titanic.info())

# Afficher des statistiques descriptives pour les colonnes numériques
print("Statistiques descriptives:")
print(df_titanic.describe())

# EXERCICE 3 : Identifier les types de données dans l'ensemble de données Titanic

# Sélectionner les colonnes qualitatives (généralement de type 'object')
qualitative_columns = df_titanic.select_dtypes(include=['object']).columns

# Sélectionner les colonnes quantitatives (typiquement int64 ou float64)
quantitative_columns = df_titanic.select_dtypes(include=['int64', 'float64']).columns

print("Colonnes qualitatives :", qualitative_columns)
print("Colonnes quantitatives :", quantitative_columns)


# EXERCICE 4 : Charger et explorer le dataset Iris

# Charger le dataset Iris 
df_iris = pd.read_csv('iris.csv')

# Afficher les 5 premières lignes pour visualiser la structure du dataset
print("Premières lignes du dataset Iris:")
print(df_iris.head())

# Identifier les colonnes qualitatives et quantitatives dans le dataset Iris
qualitative_columns_iris = df_iris.select_dtypes(include=['object']).columns
quantitative_columns_iris = df_iris.select_dtypes(include=['int64', 'float64']).columns

print("Colonnes qualitatives dans Iris :", qualitative_columns_iris)
print("Colonnes quantitatives dans Iris :", quantitative_columns_iris)


# EXERCICE 5 : Analyse de données de base sur le dataset Titanic (par exemple pour la colonne 'Fare')
!pip install -U seaborn

import matplotlib.pyplot as plt

import seaborn as sns

# Choisir une colonne quantitative, ici 'Fare'
col = 'Fare'

# Calculer la moyenne, la médiane et le mode pour la colonne 'Fare'
moyenne = df_titanic[col].mean()
mediane = df_titanic[col].median()
mode = df_titanic[col].mode()[0]  # mode() peut retourner plusieurs valeurs, on prend la première

print(f"Moyenne de {col} :", moyenne)
print(f"Médiane de {col} :", mediane)
print(f"Mode de {col} :", mode)

# Créer un histogramme pour visualiser la distribution de la colonne 'Fare'
plt.figure(figsize=(8, 5))
sns.histplot(df_titanic[col], bins=20, kde=True)
plt.xlabel("Fare")
plt.ylabel("Nombre d'observations")
plt.title("Distribution de Fare dans Titanic")
plt.show()

# EXERCICE 6 : Observation des données de sommeil
# Charger l'ensemble de données "Combien d'heures de sommeil les Américains obtiennent-ils réellement ?"


df_sleep = pd.read_csv('sleep_data.csv')

# Afficher les 5 premières lignes du dataset de sommeil
print("Premières lignes du dataset de sommeil:")
print(df_sleep.head())

# Afficher la liste des colonnes disponibles
print("Colonnes du dataset de sommeil:")
print(df_sleep.columns)

if 'Region' in df_sleep.columns:
    avg_sleep_by_region = df_sleep.groupby('Region')['Hours'].mean()
    print("Moyenne des heures de sommeil par région:")
    print(avg_sleep_by_region)
else:
    print("La colonne 'Region' n'est pas présente dans ce dataset.")
    


```
