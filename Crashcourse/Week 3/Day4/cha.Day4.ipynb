```python
df['blue'] = df['blue'].astype(int)
df['dual_sim'] = df['dual_sim'].astype(int)
df['four_g'] = df['four_g'].astype(int)
df['three_g'] = df['three_g'].astype(int)
df['touch_screen'] = df['touch_screen'].astype(int)
df['wifi'] = df['wifi'].astype(int)
```

```python

mean_price = df['price_range'].mean()
median_price = df['price_range'].median()
mode_price = df['price_range'].mode()[0]

print(f"Moyenne des prix : {mean_price}")
print(f"MÃ©diane des prix : {median_price}")
print(f"Mode des prix : {mode_price}")
```

```python
std_price = df['price_range'].std()
variance_price = df['price_range'].var()
range_price = df['price_range'].max() - df['price_range'].min()

print(f"Ã‰cart-type des prix : {std_price}")
print(f"Variance des prix : {variance_price}")
print(f"Plage des prix : {range_price}")
```

```python
# Lâ€™Ã©cart-type (1.118) et la variance (1.25) quantifient la dispersion des 4 catÃ©gories autour de la moyenne 1.5. Cela confirme que les observations sont globalement Ã©talÃ©es sur lâ€™ensemble des catÃ©gories (0, 1, 2, 3).
```

```python
skew_price = df['price_range'].skew()
kurtosis_price = df['price_range'].kurt()

print(f"AsymÃ©trie des prix : {skew_price}")
print(f"Aplatissement des prix : {kurtosis_price}")
```

```python
# AsymÃ©trie (skew) = 0.0
Une asymÃ©trie nulle indique une distribution parfaitement symÃ©trique. Ici, comme on a des classes discrÃ¨tes (0, 1, 2, 3) de maniÃ¨re presque Ã©gale, la symÃ©trie se comprend bien.
Aplatissement (kurtosis) = -1.3604
# Une kurtosis nÃ©gative (infÃ©rieure Ã  0) indique une distribution plus Â« plate Â» quâ€™une loi normale. Cela signifie que les valeurs se concentrent moins autour de la moyenne et sont plus Ã©talÃ©es (ce qui est logique avec seulement 4 catÃ©gories rÃ©parties uniformÃ©ment).
```

```python
from scipy import stats

# Test de comparaison de la moyenne des prix pour les tÃ©lÃ©phones Ã  double SIM vs non-double SIM
sim_price = df[df['dual_sim'] == 1]['price_range']
non_sim_price = df[df['dual_sim'] == 0]['price_range']

t_stat, p_value = stats.ttest_ind(sim_price, non_sim_price)

print(f"Test t pour comparer les prix : t-statistic = {t_stat}, p-value = {p_value}")
```

```python
# le fait quâ€™un tÃ©lÃ©phone ait une double SIM ou non nâ€™influence pas significativement la gamme de prix dans ce dataset
```
