```python
import numpy as np 
np.random.seed(42)  # Fixer le seed pour avoir des résultats reproductibles
matrice = np.random.randint(1, 100, (5, 5))  # Matrice 5x5 avec valeurs entre 1 et 100
print("Matrice originale :\n", matrice)
```

```python


# 2️⃣ Calcul des valeurs propres et vecteurs propres
valeurs_propres, vecteurs_propres = np.linalg.eig(matrice)
print("\nValeurs propres :\n", valeurs_propres)
print("\nVecteurs propres :\n", vecteurs_propres)
```

```python
# 3️⃣ Normalisation de la matrice
# La normalisation se fait généralement avec : (x - min) / (max - min)
min_val = np.min(matrice)
max_val = np.max(matrice)

matrice_normalisee = (matrice - min_val) / (max_val - min_val)
print("\nMatrice normalisée :\n", matrice_normalisee)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats  # Pour calculer l'asymétrie et la kurtosis

# Génération de 1000 valeurs suivant une distribution normale
moyenne = 50    # Valeur moyenne (centre de la cloche)
ecart_type = 10 # Dispersion des valeurs
taille = 1000   # Nombre de valeurs générées

donnees = np.random.normal(loc=moyenne, scale=ecart_type, size=taille)

# Affichage de l'histogramme
plt.hist(donnees, bins=30, color='skyblue', edgecolor='black', density=True)
plt.xlabel('Valeurs')
plt.ylabel('Densité')
plt.title('Histogramme de la distribution normale')
plt.grid()
plt.show()
```
