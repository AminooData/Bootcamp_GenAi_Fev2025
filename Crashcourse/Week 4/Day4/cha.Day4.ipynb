```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_csv('train.csv', encoding='ISO-8859-1')
df.head()
```

```python
df['blue'] = df['blue'].astype(int)
df['dual_sim'] = df['dual_sim'].astype(int)
df['four_g'] = df['four_g'].astype(int)
df['three_g'] = df['three_g'].astype(int)
df['touch_screen'] = df['touch_screen'].astype(int)
df['wifi'] = df['wifi'].astype(int)
```

```python

mean_price = df['price_range'].mean()
median_price = df['price_range'].median()
mode_price = df['price_range'].mode()[0]

print(f"Moyenne des prix : {mean_price}")
print(f"Médiane des prix : {median_price}")
print(f"Mode des prix : {mode_price}")
```

```python
std_price = df['price_range'].std()
variance_price = df['price_range'].var()
range_price = df['price_range'].max() - df['price_range'].min()

print(f"Écart-type des prix : {std_price}")
print(f"Variance des prix : {variance_price}")
print(f"Plage des prix : {range_price}")
```

```python
# L’écart-type (1.118) et la variance (1.25) quantifient la dispersion des 4 catégories autour de la moyenne 1.5. Cela confirme que les observations sont globalement étalées sur l’ensemble des catégories (0, 1, 2, 3).
```

```python
skew_price = df['price_range'].skew()
kurtosis_price = df['price_range'].kurt()

print(f"Asymétrie des prix : {skew_price}")
print(f"Aplatissement des prix : {kurtosis_price}")
```

```python
# Asymétrie (skew) = 0.0
Une asymétrie nulle indique une distribution parfaitement symétrique. Ici, comme on a des classes discrètes (0, 1, 2, 3) de manière presque égale, la symétrie se comprend bien.
Aplatissement (kurtosis) = -1.3604
# Une kurtosis négative (inférieure à 0) indique une distribution plus « plate » qu’une loi normale. Cela signifie que les valeurs se concentrent moins autour de la moyenne et sont plus étalées (ce qui est logique avec seulement 4 catégories réparties uniformément).
```

```python
from scipy import stats

# Test de comparaison de la moyenne des prix pour les téléphones à double SIM vs non-double SIM
sim_price = df[df['dual_sim'] == 1]['price_range']
non_sim_price = df[df['dual_sim'] == 0]['price_range']

t_stat, p_value = stats.ttest_ind(sim_price, non_sim_price)

print(f"Test t pour comparer les prix : t-statistic = {t_stat}, p-value = {p_value}")
```

```python
# le fait qu’un téléphone ait une double SIM ou non n’influence pas significativement la gamme de prix dans ce dataset
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10,6))
sns.histplot(df['price_range'], kde=True, bins=30)
plt.title('Distribution des prix des téléphones')
plt.xlabel('Gamme de prix')
plt.ylabel('Fréquence')
plt.show()
```

```python
# L’histogramme confirme que chaque classe (0, 1, 2, 3) est bien représentée, rendant la distribution relativement « plate » et symétrique.
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='n_cores', y='price_range', data=df)
plt.title('Box plot des prix en fonction du nombre de cœurs')
plt.xlabel('Nombre de cœurs')
plt.ylabel('Gamme de prix')
plt.show()
```

```python
# # Tendance globale : Les téléphones ayant 5, 6 ou 7 cœurs ont souvent des gammes de prix plus élevées (2 ou 3) que ceux ayant 1 ou 2 cœurs (plutôt 0 ou 1).
# Variabilité : Les boîtes (box) et les moustaches (whiskers) montrent qu’il existe des téléphones à 4 ou 5 cœurs classés dans des gammes de prix plus basses ou plus hautes. Cela signifie que le nombre de cœurs n’est pas le seul facteur déterminant le prix, mais qu’il y a une corrélation partielle.
```
