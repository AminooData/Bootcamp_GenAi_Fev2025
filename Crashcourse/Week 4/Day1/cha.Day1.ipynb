```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Définition des villes et des mois
villes = ["Paris", "Londres", "New York", "Tokyo", "Moscou", "Sydney", "Dubai", "Toronto", "Pékin", "Rio"]
mois = ["Jan", "Fév", "Mar", "Avr", "Mai", "Juin", "Juil", "Août", "Sep", "Oct", "Nov", "Déc"]

# Générer des températures aléatoires entre -5°C et 35°C
temp_data = np.random.uniform(-5, 35, (10, 12))

# Convertir en DataFrame Pandas
df_temp = pd.DataFrame(temp_data, index=villes, columns=mois)

# Affichage du DataFrame
print(df_temp.head())
```

```python
# Calcul de la température moyenne annuelle
df_temp["Moyenne Annuelle"] = df_temp.mean(axis=1)

# Trouver la ville la plus chaude et la plus froide
ville_plus_chaude = df_temp["Moyenne Annuelle"].idxmax()
ville_plus_froide = df_temp["Moyenne Annuelle"].idxmin()

# Affichage des résultats
print(f" Ville la plus chaude : {ville_plus_chaude} ({df_temp.loc[ville_plus_chaude, 'Moyenne Annuelle']:.2f}°C)")
print(f" Ville la plus froide : {ville_plus_froide} ({df_temp.loc[ville_plus_froide, 'Moyenne Annuelle']:.2f}°C)")
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df_temp.iloc[:, :-1], annot=True, cmap="coolwarm", fmt=".1f", linewidths=0.5)
plt.title(" Températures Moyennes Mensuelles par Ville")
plt.xlabel("Mois")
plt.ylabel("Villes")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=df_temp.index, y=df_temp["Moyenne Annuelle"], palette="coolwarm")
plt.xticks(rotation=45)
plt.title("🌍 Température Moyenne Annuelle par Ville")
plt.xlabel("Villes")
plt.ylabel("Température Moyenne (°C)")
plt.show()
```
