```python
# # Exercice 1 : Petit quizz
# Quelle est la principale différence entre l’apprentissage automatique traditionnel et l’apprentissage profond ?
#  L’apprentissage automatique traditionnel nécessite une extraction manuelle des caractéristiques, tandis que l’apprentissage profond apprend automatiquement les représentations pertinentes à partir des données grâce aux réseaux de neurones profonds.

# Comment les réseaux de neurones artificiels (RNA) imitent-ils le cerveau humain ?
# →Ils imitent le cerveau en utilisant des neurones artificiels interconnectés qui traitent et transmettent l’information de manière hiérarchique, comme les neurones biologiques.

# Pourquoi l’apprentissage profond est-il plus performant sur de grands ensembles de données que l’apprentissage automatique traditionnel ?
# Parce qu’il peut apprendre des représentations complexes directement à partir des données sans nécessiter d’ingénierie des caractéristiques, et les réseaux profonds deviennent plus précis avec plus de données.

# Quels sont les défis de l’apprentissage profond et comment peuvent-ils être relevés ?
# Les principaux défis incluent le besoin de grandes quantités de données, la puissance de calcul élevée et la difficulté d’interprétation. Ils peuvent être relevés avec des techniques comme le transfert d’apprentissage, l’optimisation des modèles et l’explicabilité des réseaux.

# Qu’est-ce que l’ingénierie des fonctionnalités et pourquoi n’est-elle pas nécessaire dans l’apprentissage profond ?
# L’ingénierie des fonctionnalités consiste à sélectionner et transformer les données en caractéristiques exploitables par un modèle. Dans l’apprentissage profond, elle est souvent inutile car le réseau extrait lui-même les caractéristiques les plus pertinentes.

# Quel rôle jouent les couches cachées dans un modèle d’apprentissage profond ?
# Elles permettent d’extraire des représentations abstraites des données en effectuant des transformations successives, rendant possible la capture de relations complexes.

# Dans un réseau neuronal artificiel (ANN), quelle est la fonction d'une fonction d'activation ?
# Elle introduit de la non-linéarité dans le modèle, permettant au réseau d’apprendre des relations complexes et d’améliorer la capacité de généralisation.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def perceptron_decision(temp, rain):
    weight_temp = 0.6
    weight_rain = 0.4
    bias = 2
    weighted_sum = (temp * weight_temp) + (rain * weight_rain) + bias
    return 1 if weighted_sum > 20 else 0

# Test des cas donnés
print("Cas 1 (70°F, pas de pluie) :", perceptron_decision(70, 0))  # Attendu : 1 (sortir)
print("Cas 2 (50°F, pluie) :", perceptron_decision(50, 1))  # Attendu : 0 (rester)

# Un perceptron prend des entrées pondérées, les additionne et applique une fonction d'activation (Step Function) pour prendre une décision.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return np.where(x > 0, 1, 0)

def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))

def relu_function(x):
    return np.maximum(0, x)

x_values = np.linspace(-10, 10, 100)

plt.figure(figsize=(12, 4))

# Step Function Plot
plt.subplot(1, 3, 1)
plt.plot(x_values, step_function(x_values), label="Step Function", color='blue')
plt.title("Step Function")
plt.grid(True)

# Sigmoid Function Plot
plt.subplot(1, 3, 2)
plt.plot(x_values, sigmoid_function(x_values), label="Sigmoid Function", color='red')
plt.title("Sigmoid Function")
plt.grid(True)

# ReLU Function Plot
plt.subplot(1, 3, 3)
plt.plot(x_values, relu_function(x_values), label="ReLU Function", color='green')
plt.title("ReLU Function")
plt.grid(True)

plt.show()

# Pourquoi ReLU est-il couramment utilisé dans les modèles d’apprentissage profond ?

# ReLU (Rectified Linear Unit) est très efficace car il introduit de la non-linéarité tout en évitant le problème de la disparition du gradient (vanishing gradient).
# Contrairement à la Sigmoid ou la Tangente Hyperbolique, ReLU ne sature pas pour les grandes valeurs positives, ce qui accélère l'entraînement.
# Son calcul est simple : si 𝑥 > 0 x>0, il garde 𝑥 sinon il met  0. Cela permet des mises à jour plus efficaces lors de la descente de gradient.

# 2️ Pourquoi Sigmoid pourrait-il être un bon choix pour les tâches de classification binaire ?

# La Sigmoid produit une sortie entre 0 et 1, ce qui est idéal pour interpréter une probabilité.
# Si la sortie 𝑦  est proche de 0, le modèle prédit la classe 0, et si elle est proche de 1, il prédit la classe 1.
# C'est particulièrement utile pour les problèmes où la sortie doit être une probabilité, comme la classification d'un email en spam ou non spam.

# 3️ Quelles sont les limites de la fonction Step par rapport aux autres ?

# La fonction Step est brusque : elle renvoie uniquement 0 ou 1, ce qui empêche l’apprentissage progressif d’un modèle.
# Elle ne permet pas le calcul du gradient, donc elle n'est pas utilisée pour entraîner des réseaux de neurones via la descente de gradient.
# Contrairement à ReLU et Sigmoid, elle ne permet pas de gérer des valeurs continues, ce qui la rend peu utile pour les modèles complexes.
```

```python
def relu(x):
    return max(0, x)

def predict_house_price(x1, x2):
    w1, w2, b = 0.5, 0.7, 50000
    z = (x1 * w1) + (x2 * w2) + b
    return relu(z)

predicted_price = predict_house_price(2000, 3)
print("Prix prédit de la maison :", predicted_price)

# Le modèle prédit que le prix de la maison est de 51 002.1 (unité non précisée, mais probablement en dollars).
```

```python
import numpy as np

# Données d'entrée (features)
x = np.array([4, 80])  # 4 heures étudiées, score précédent de 80

# Poids et biais initiaux
w = np.array([0.6, 0.3])
b = 10

# Fonction de propagation vers l'avant
def forward_propagation(x, w, b):
    return np.dot(x, w) + b  # Pas de ReLU, car c'est une tâche de régression

# Prédiction initiale
y_pred = forward_propagation(x, w, b)
y_true = 85  # Score réel

# Calcul de la perte (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2

# Calcul des gradients
grad_w = -(y_true - y_pred) * x
grad_b = -(y_true - y_pred)

# Mise à jour des poids et biais (Descente de gradient)
learning_rate = 0.01
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

# Résultats
print("Prédiction initiale :", y_pred)
print("Erreur :", loss)
print("Nouveaux poids :", w_new)
print("Nouveau biais :", b_new)

#  Pourquoi la mise à jour des poids réduit-elle l’erreur ?

# La descente de gradient ajuste les poids et le biais dans le sens opposé à l'erreur, ce qui réduit progressivement la différence entre la prédiction et la valeur réelle.
# À chaque mise à jour, les poids sont modifiés de manière à minimiser la fonction de perte (ici, l'erreur quadratique moyenne MSE).
# # Cela permet au modèle de mieux s’adapter aux données après plusieurs itérations.
```

```python

w = np.array([0.8, 0.5])  # Poids plus élevés
learning_rate = 0.05  # Taux d'apprentissage plus grand

# Propagation avant
y_pred = forward_propagation(x, w, b)
loss = 0.5 * (y_true - y_pred) ** 2

# Mise à jour des poids
grad_w = -(y_true - y_pred) * x
grad_b = -(y_true - y_pred)
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

print("Nouvelle prédiction après modification :", forward_propagation(x, w_new, b_new))
print("Nouvelle perte :", 0.5 * (y_true - forward_propagation(x, w_new, b_new)) ** 2)
print("Nouveaux poids :", w_new)
print("Nouveau biais :", b_new)
```
