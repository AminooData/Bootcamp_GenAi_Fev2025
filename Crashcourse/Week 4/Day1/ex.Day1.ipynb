```python
# # Exercice 1 : Petit quizz
# Quelle est la principale diff√©rence entre l‚Äôapprentissage automatique traditionnel et l‚Äôapprentissage profond ?
#  L‚Äôapprentissage automatique traditionnel n√©cessite une extraction manuelle des caract√©ristiques, tandis que l‚Äôapprentissage profond apprend automatiquement les repr√©sentations pertinentes √† partir des donn√©es gr√¢ce aux r√©seaux de neurones profonds.

# Comment les r√©seaux de neurones artificiels (RNA) imitent-ils le cerveau humain ?
# ‚ÜíIls imitent le cerveau en utilisant des neurones artificiels interconnect√©s qui traitent et transmettent l‚Äôinformation de mani√®re hi√©rarchique, comme les neurones biologiques.

# Pourquoi l‚Äôapprentissage profond est-il plus performant sur de grands ensembles de donn√©es que l‚Äôapprentissage automatique traditionnel ?
# Parce qu‚Äôil peut apprendre des repr√©sentations complexes directement √† partir des donn√©es sans n√©cessiter d‚Äôing√©nierie des caract√©ristiques, et les r√©seaux profonds deviennent plus pr√©cis avec plus de donn√©es.

# Quels sont les d√©fis de l‚Äôapprentissage profond et comment peuvent-ils √™tre relev√©s ?
# Les principaux d√©fis incluent le besoin de grandes quantit√©s de donn√©es, la puissance de calcul √©lev√©e et la difficult√© d‚Äôinterpr√©tation. Ils peuvent √™tre relev√©s avec des techniques comme le transfert d‚Äôapprentissage, l‚Äôoptimisation des mod√®les et l‚Äôexplicabilit√© des r√©seaux.

# Qu‚Äôest-ce que l‚Äôing√©nierie des fonctionnalit√©s et pourquoi n‚Äôest-elle pas n√©cessaire dans l‚Äôapprentissage profond ?
# L‚Äôing√©nierie des fonctionnalit√©s consiste √† s√©lectionner et transformer les donn√©es en caract√©ristiques exploitables par un mod√®le. Dans l‚Äôapprentissage profond, elle est souvent inutile car le r√©seau extrait lui-m√™me les caract√©ristiques les plus pertinentes.

# Quel r√¥le jouent les couches cach√©es dans un mod√®le d‚Äôapprentissage profond ?
# Elles permettent d‚Äôextraire des repr√©sentations abstraites des donn√©es en effectuant des transformations successives, rendant possible la capture de relations complexes.

# Dans un r√©seau neuronal artificiel (ANN), quelle est la fonction d'une fonction d'activation ?
# Elle introduit de la non-lin√©arit√© dans le mod√®le, permettant au r√©seau d‚Äôapprendre des relations complexes et d‚Äôam√©liorer la capacit√© de g√©n√©ralisation.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def perceptron_decision(temp, rain):
    weight_temp = 0.6
    weight_rain = 0.4
    bias = 2
    weighted_sum = (temp * weight_temp) + (rain * weight_rain) + bias
    return 1 if weighted_sum > 20 else 0

# Test des cas donn√©s
print("Cas 1 (70¬∞F, pas de pluie) :", perceptron_decision(70, 0))  # Attendu : 1 (sortir)
print("Cas 2 (50¬∞F, pluie) :", perceptron_decision(50, 1))  # Attendu : 0 (rester)

# Un perceptron prend des entr√©es pond√©r√©es, les additionne et applique une fonction d'activation (Step Function) pour prendre une d√©cision.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return np.where(x > 0, 1, 0)

def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))

def relu_function(x):
    return np.maximum(0, x)

x_values = np.linspace(-10, 10, 100)

plt.figure(figsize=(12, 4))

# Step Function Plot
plt.subplot(1, 3, 1)
plt.plot(x_values, step_function(x_values), label="Step Function", color='blue')
plt.title("Step Function")
plt.grid(True)

# Sigmoid Function Plot
plt.subplot(1, 3, 2)
plt.plot(x_values, sigmoid_function(x_values), label="Sigmoid Function", color='red')
plt.title("Sigmoid Function")
plt.grid(True)

# ReLU Function Plot
plt.subplot(1, 3, 3)
plt.plot(x_values, relu_function(x_values), label="ReLU Function", color='green')
plt.title("ReLU Function")
plt.grid(True)

plt.show()

# Pourquoi ReLU est-il couramment utilis√© dans les mod√®les d‚Äôapprentissage profond ?

# ReLU (Rectified Linear Unit) est tr√®s efficace car il introduit de la non-lin√©arit√© tout en √©vitant le probl√®me de la disparition du gradient (vanishing gradient).
# Contrairement √† la Sigmoid ou la Tangente Hyperbolique, ReLU ne sature pas pour les grandes valeurs positives, ce qui acc√©l√®re l'entra√Ænement.
# Son calcul est simple : si ùë• > 0 x>0, il garde ùë• sinon il met  0. Cela permet des mises √† jour plus efficaces lors de la descente de gradient.

# 2Ô∏è Pourquoi Sigmoid pourrait-il √™tre un bon choix pour les t√¢ches de classification binaire ?

# La Sigmoid produit une sortie entre 0 et 1, ce qui est id√©al pour interpr√©ter une probabilit√©.
# Si la sortie ùë¶  est proche de 0, le mod√®le pr√©dit la classe 0, et si elle est proche de 1, il pr√©dit la classe 1.
# C'est particuli√®rement utile pour les probl√®mes o√π la sortie doit √™tre une probabilit√©, comme la classification d'un email en spam ou non spam.

# 3Ô∏è Quelles sont les limites de la fonction Step par rapport aux autres ?

# La fonction Step est brusque : elle renvoie uniquement 0 ou 1, ce qui emp√™che l‚Äôapprentissage progressif d‚Äôun mod√®le.
# Elle ne permet pas le calcul du gradient, donc elle n'est pas utilis√©e pour entra√Æner des r√©seaux de neurones via la descente de gradient.
# Contrairement √† ReLU et Sigmoid, elle ne permet pas de g√©rer des valeurs continues, ce qui la rend peu utile pour les mod√®les complexes.
```

```python
def relu(x):
    return max(0, x)

def predict_house_price(x1, x2):
    w1, w2, b = 0.5, 0.7, 50000
    z = (x1 * w1) + (x2 * w2) + b
    return relu(z)

predicted_price = predict_house_price(2000, 3)
print("Prix pr√©dit de la maison :", predicted_price)

# Le mod√®le pr√©dit que le prix de la maison est de 51 002.1 (unit√© non pr√©cis√©e, mais probablement en dollars).
```

```python
import numpy as np

# Donn√©es d'entr√©e (features)
x = np.array([4, 80])  # 4 heures √©tudi√©es, score pr√©c√©dent de 80

# Poids et biais initiaux
w = np.array([0.6, 0.3])
b = 10

# Fonction de propagation vers l'avant
def forward_propagation(x, w, b):
    return np.dot(x, w) + b  # Pas de ReLU, car c'est une t√¢che de r√©gression

# Pr√©diction initiale
y_pred = forward_propagation(x, w, b)
y_true = 85  # Score r√©el

# Calcul de la perte (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2

# Calcul des gradients
grad_w = -(y_true - y_pred) * x
grad_b = -(y_true - y_pred)

# Mise √† jour des poids et biais (Descente de gradient)
learning_rate = 0.01
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

# R√©sultats
print("Pr√©diction initiale :", y_pred)
print("Erreur :", loss)
print("Nouveaux poids :", w_new)
print("Nouveau biais :", b_new)

#  Pourquoi la mise √† jour des poids r√©duit-elle l‚Äôerreur ?

# La descente de gradient ajuste les poids et le biais dans le sens oppos√© √† l'erreur, ce qui r√©duit progressivement la diff√©rence entre la pr√©diction et la valeur r√©elle.
# √Ä chaque mise √† jour, les poids sont modifi√©s de mani√®re √† minimiser la fonction de perte (ici, l'erreur quadratique moyenne MSE).
# # Cela permet au mod√®le de mieux s‚Äôadapter aux donn√©es apr√®s plusieurs it√©rations.
```

```python

w = np.array([0.8, 0.5])  # Poids plus √©lev√©s
learning_rate = 0.05  # Taux d'apprentissage plus grand

# Propagation avant
y_pred = forward_propagation(x, w, b)
loss = 0.5 * (y_true - y_pred) ** 2

# Mise √† jour des poids
grad_w = -(y_true - y_pred) * x
grad_b = -(y_true - y_pred)
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

print("Nouvelle pr√©diction apr√®s modification :", forward_propagation(x, w_new, b_new))
print("Nouvelle perte :", 0.5 * (y_true - forward_propagation(x, w_new, b_new)) ** 2)
print("Nouveaux poids :", w_new)
print("Nouveau biais :", b_new)
```
